```python
from bs4 import BeautifulSoup
import urllib.request as req
import csv
```

```python
url = "http://bbs.kakaku.com/bbs/-/CategoryCD=7010/SortRule=1/ResView=all/#tab"
# urlopen()でデータを取得
res = req.urlopen(url)
```

```python
# BeautifulSoup()で解析
soup = BeautifulSoup(res, 'html.parser')
```

```python
# 任意のデータを抽出
title1 = soup.find("h1").string
print("title = ", title1)

p_list = soup.find_all("p")

for p in p_list:
    print(p.get_text())
```

```python
# 任意のデータを抽出

kakaku_kuchikomi = []

for p in p_list:
    kakaku_kuchikomi.append(p.get_text())

with open('kakaku_kuchikomi.csv', 'w', encoding='utf-8' ) as f:
    writer = csv.writer(f)
    writer.writerows(kakaku_kuchikomi)
```
